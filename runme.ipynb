## Algorithm defined in reference paper to find communities
1. Create an edge labelled directed multigraph
2. Convert above graph to edge weighted undirected graph
3. Generate embeddings using node2vec algorithm
4. Apply DBSCAN to identify clusters of nodes connected densely together

```python
# Flow 

# Step 0: Import data
path = 'data/dataset.csv'
data = import_data(path)

# Step 1: Create an edge labelled directed multigraph
multi_graph = create_multi_graph(data)

# Step 2: Convert above graph to edge weighted undirected graph
graph = create_undirected_graph(multi_graph)

# Step 3: Generate embeddings using node2vec
embeddings = generate_embeddings(graph)

# Step 4: Apply DBSCAN to identify clusters of nodes connected densely together
tsne, clusters = dbs(embeddings, eps = 1.5, min_samples = 3)

# Step 5: Plot the results
cluster_plot(tsne, clusters)
```
